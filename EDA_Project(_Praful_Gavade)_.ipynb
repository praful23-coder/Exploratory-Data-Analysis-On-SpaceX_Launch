#  Bivariate Analysis (Relationship Between Two Columns)

```python
# Caluating the Satellite weight in the Rocket


plt.figure(figsize=(10,10))
sns.lineplot(x='Flight Number', y='Payload Mass (kg)', data=SpaceX_Launch, marker='o', color='blue')
plt.title('Payload Mass by Flight Number')
plt.show()
```

```python
# Realtion of Launch site and Landing outcome
plt.figure(figsize=(20,20))
sns.catplot( x='Mission Outcome',y='Launch Site',hue='Landing Outcome',kind='bar',data=SpaceX_Launch)
sns.color_palette("magma", as_cmap=True)
plt.title('Landing Outcome by Launch Site')
plt.ylabel('Number of Landing')
plt.xlabel('Launch Site')
plt.xticks(rotation=90)
plt.show()
```

```python
# calulating  the date
SpaceX_Launch['Date'] = pd.to_datetime(SpaceX_Launch['Date'], errors='coerce')
SpaceX_Launch['Time'] = SpaceX_Launch['Date'].dt.year

# Calculating percentage of successful missions each year
success1 = SpaceX_Launch.groupby('Time')['Mission Outcome'].apply(lambda x: (x == 'Success').mean() * 100)
success1



# Making the graph
plt.figure(figsize=(10, 5))
success1.plot(marker='o', color='green')
plt.title("Mission Success Rate Per Year")
plt.xlabel("Year")
plt.ylabel("Success Rate (%)")
plt.show()

```

```python
# Analysis the launch site with time on launch site

plt.figure(figsize=(15,10))
sns.pointplot(data=SpaceX_Launch, x='Landing Outcome', y='Time (UTC)', hue='Landing Outcome')
plt.title("Landing Outcome by Launch Hour")
plt.xlabel("Hour of Launch (UTC)")
plt.ylabel("Landing Outcome")
plt.xticks(rotation=90)
plt.show()

```

```python
# Booster Version and Mission Outcome success and Failure in bar plot
plt.figure(figsize=(15,10))
sns.barplot(data=SpaceX_Launch, x='Mission Outcome', y='Booster Version')
plt.title("Mission Outcome by Booster Version")
plt.xlabel("Booster Version")
plt.ylabel("Mission Outcome")
plt.show()
```

```python
# Success and Failure rate of time with mission outcome

plt.figure(figsize=(15, 5))
sns.pointplot(data=SpaceX_Launch, x='Time (UTC)', y='Mission Outcome', color='skyblue')
plt.xticks(rotation=90)
plt.title('Mission Success Rate by Launch Time')
plt.ylabel('Success Rate (%)')
plt.xlabel('Time (UTC)')
plt.show()

```

5.Payload Mass:-
* Payloads greatly, but heavier payloads did not directly lead to more mission failures — showing SpaceX stability in handling diverse missions.
